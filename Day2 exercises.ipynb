```python
df1 = pd.read_csv("flights.csv")
df1.head()
```

```python
# Identify the shape of the dataset (number of rows and columns)
print(df1.shape)
```

```python
# Display the column names, data types, and non-null values
df1.info()
```

```python
# View the first and last few rows of the dataset.
df1.head()
```

```python
df1.tail()
```

```python
# Check for missing values in the dataset.
df1.isnull().sum()
```

```python
# Generate summary statistics for numerical columns, including mean, median, standard deviation, and percentiles.
df1.describe()
```

```python
# Extract records where flights were delayed by more than 15 minutes or were canceled
filtered_df = df1[(df1['DepDel15'] > 15) | (df1['Cancelled'] == True)]
filtered_df.head()
```

```python
# Analyze the average delay for each airline (Carrier) or by airport.
average_delay_by_carrier = df1.groupby('Carrier')[['DepDelay', 'ArrDelay']].mean()
average_delay_by_carrier
```

```python
# Sort the dataset by DepDelay in descending order to find the most delayed flights.
sorted_df = df1.sort_values(by='DepDelay', ascending=False)
sorted_df
```
