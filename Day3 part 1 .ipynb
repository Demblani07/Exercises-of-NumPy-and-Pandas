```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("flights.csv")
df.head()
```

Part 1 day 3 Quick hands on exercise
 
Plot subplot for using Matplotlib lib Visualization module
 
Flights:
I - 
Month versus Departure
Month versus Arrivals
II - 
MOnth versus Delaydepartue
Mnth versus DelayArrival

```python
departures_per_month = df.groupby('Month')['DepDelay'].count()
arrivals_per_month = df.groupby('Month')['ArrDelay'].count()
avg_departure_delay = df.groupby('Month')['DepDelay'].mean()
avg_arrival_delay = df.groupby('Month')['ArrDelay'].mean()

fig, axes = plt.subplots(2, 2, figsize=(12, 8))

axes[0, 0].bar(departures_per_month.index, departures_per_month.values, color='skyblue')
axes[0, 0].set_title('Month vs Departures')
axes[0, 0].set_xlabel('Month')
axes[0, 0].set_ylabel('Number of Departures')

axes[0, 1].bar(arrivals_per_month.index, arrivals_per_month.values, color='orange')
axes[0, 1].set_title('Month vs Arrivals')
axes[0, 1].set_xlabel('Month')
axes[0, 1].set_ylabel('Number of Arrivals')

axes[1, 0].plot(avg_departure_delay.index, avg_departure_delay.values, marker='o', color='green')
axes[1, 0].set_title('Month vs Avg Departure Delay')
axes[1, 0].set_xlabel('Month')
axes[1, 0].set_ylabel('Avg Departure Delay (minutes)')

axes[1, 1].plot(avg_arrival_delay.index, avg_arrival_delay.values, marker='o', color='red')
axes[1, 1].set_title('Month vs Avg Arrival Delay')
axes[1, 1].set_xlabel('Month')
axes[1, 1].set_ylabel('Avg Arrival Delay (minutes)')

plt.tight_layout()
plt.show()
```
